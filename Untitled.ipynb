```python
import seaborn as sns

import cv2
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
disease_list = ['COVID', 'non-COVID']

train_dir = data_dir = 'dataset'

train_data = []

for index, sp in enumerate(disease_list):
    for file in os.listdir(os.path.join(train_dir, sp)):
        train_data.append([sp + "/" + file, index, sp])
        
train = pd.DataFrame(train_data, columns = ['File', 'ID','Disease'])
train
```

```python
seed = 30
IMAGE_SIZE = 224
train = train.sample(frac = 1, replace=False, random_state = seed) 
train = train.reset_index(drop = True) # Reset indices (row numbers)

sns.countplot(x = "Disease", data = train).set_title("Frequency Histogram of Disease")
```

```python
def plot_defects(defect_types, rows, cols):
    fig, ax = plt.subplots(rows, cols, figsize=(12, 12))
    defect_files = train['File'][train['Disease'] == defect_types].values
    
    n = 0
    fig.suptitle(defect_types, fontsize = 22, color = "black")
    for i in range(rows):
        for j in range(cols):
            image_path = os.path.join(data_dir, defect_files[n])
            ax[i, j].set_xticks([])
            ax[i, j].set_yticks([])
            ax[i, j].imshow(cv2.imread(image_path))
            n += 1
            
# Displays first n images of class from training set
plot_defects('COVID', 2, 3)
plot_defects('non-COVID', 2, 3)
```
